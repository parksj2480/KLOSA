# 결측치 제거하기

```python
for i in range(8):
    num=i+1
    df_main=pd.read_csv('str0{}.csv'.format(num),encoding='cp949')

    #Regular Expression을 사용하여 01~07까지 적용되도록 함.
    df=df_main.loc[:,df_main.columns.str.contains(r'pid|A002_age$|gender1$|region1$|region3$|hhsize$|edu$|edu_s$|martial$|C083$|C075$|chronic_a|chronic_b|chronic_c|chronic_d|chronic_e|chronic_f|chronic_g|chronic_h|chronic_i|C105$|C107$|bmi$|smoke$|alc$|adl$|iadl$|C108$|C152$|C144$|C056$|present_labor$|mmse$|mmseg$')]
    df.shape

    # 결측치가 -9 혹은 None 으로 표시되어있으므로 None 으로 표시되게끔 통일함.
    df=df.replace(-9,None)
    df=df.replace(-8,None)

    # chronic_d 는 만성폐질환 진단여부 1 : 예 5 : 아니오 이므로 무응답=아니오로 대체한다.
    try:
        df.loc[df[df.columns[df.columns.str.contains(r'chronic_a$')]].isna().squeeze(),df.columns[df.columns.str.contains(r'chronic_a$')][0]]=5
        df.loc[df[df.columns[df.columns.str.contains(r'chronic_b$')]].isna().squeeze(),df.columns[df.columns.str.contains(r'chronic_b$')][0]]=5
        df.loc[df[df.columns[df.columns.str.contains(r'chronic_c$')]].isna().squeeze(),df.columns[df.columns.str.contains(r'chronic_c$')][0]]=5
        df.loc[df[df.columns[df.columns.str.contains(r'chronic_d$')]].isna().squeeze(),df.columns[df.columns.str.contains(r'chronic_d$')][0]]=5
        df.loc[df[df.columns[df.columns.str.contains(r'chronic_e$')]].isna().squeeze(),df.columns[df.columns.str.contains(r'chronic_e$')][0]]=5
        df.loc[df[df.columns[df.columns.str.contains(r'chronic_f$')]].isna().squeeze(),df.columns[df.columns.str.contains(r'chronic_f$')][0]]=5
        df.loc[df[df.columns[df.columns.str.contains(r'chronic_g$')]].isna().squeeze(),df.columns[df.columns.str.contains(r'chronic_g$')][0]]=5
        df.loc[df[df.columns[df.columns.str.contains(r'chronic_h$')]].isna().squeeze(),df.columns[df.columns.str.contains(r'chronic_h$')][0]]=5
        df.loc[df[df.columns[df.columns.str.contains(r'chronic_i$')]].isna().squeeze(),df.columns[df.columns.str.contains(r'chronic_i$')][0]]=5
    except:
        print("No Null in chronic!")


    try:
        # Edu 교육수준 1:무학(문자해독불가) 2:무학(문자해독가능) 3:초졸 4: 중졸 최빈값 1로 대체
        df.loc[df[df.columns[df.columns.str.contains(r'edu$')]].isna().squeeze(),df.columns[df.columns.str.contains(r'edu$')][0]]=1
    except:
        None
    try:    
        # 현재 시력으로 인한 일상생활의 어려움 여부 1: 그렇다 5: 아니다
        df.loc[df[df.columns[df.columns.str.contains(r'C081$')]].isna().squeeze(),df.columns[df.columns.str.contains(r'C081$')][0]]=5
    except:
        None
    try:
        # 현재 음주중인지 1:현재음주자 2:과거음주자 3:비음주자 결측치가 하나이므로 그냥 3으로 채워넣는다
        df.loc[df[df.columns[df.columns.str.contains(r'alc$')]].isna().squeeze(),df.columns[df.columns.str.contains(r'alc$')][0]]=3
    except:
        None
    try:
        # smoke 에 대한 변수 0:비흡연자 1:과거흡연자 2:현재흡연자 가장 빈도수가 많은 비흡연자로 채워넣는다.
        df.loc[df[df.columns[df.columns.str.contains(r'smoke$')]].isna().squeeze(),df.columns[df.columns.str.contains(r'smoke$')][0]]=0
    except:
        None
    try:    
        # 우울증 최빈값 1로 대체
        df.loc[df[df.columns[df.columns.str.contains(r'C144$')]].isna().squeeze(),df.columns[df.columns.str.contains(r'C144$')][0]]=1
    except:
        None
    try:    
        # 시력 최빈값 3로 대체
        df.loc[df[df.columns[df.columns.str.contains(r'C075$')]].isna().squeeze(),df.columns[df.columns.str.contains(r'C075$')][0]]=3
    except:
        print("No null in else")



    # mmse 점수가 없는 사람은 버리는 것을 선택한다. w07에서는 342명의 데이터가 소실된다.
    try:
        df.dropna(axis=0,subset=df.columns[df.columns.str.contains(r'mmse$')],inplace=True)
        print("MMSE Null eliminated!")
    except:
        print("No mmse Null found!")

    # 요일에 대한 인지력 1:정답 5:오답
    # df.w07C402.value_counts()

    df=df.set_index('pid')
    df.head()

    df.to_csv("subdata0{}.csv".format(num))
```

```python
df.drop(columns=['w01edu','w02edu','w03edu','w04edu','w05edu','w06edu','w07edu'],inplace=True)
```

```python
# 남녀 중앙값으로 몸무게와 키를 채워넣는다.
women_height=df[df[df.columns[df.columns.str.contains(r'gender1$')][0]]==5][df.columns[df.columns.str.contains(r'C107$')][0]].median()
women_weight=df[df[df.columns[df.columns.str.contains(r'gender1$')][0]]==5][df.columns[df.columns.str.contains(r'C105$')][0]].median()
men_height=df[df[df.columns[df.columns.str.contains(r'gender1$')][0]]==1][df.columns[df.columns.str.contains(r'C107$')][0]].median()
men_weight=df[df[df.columns[df.columns.str.contains(r'gender1$')][0]]==1][df.columns[df.columns.str.contains(r'C105$')][0]].median()

gender_str=df.columns[df.columns.str.contains(r'gender1$')][0]
height_str=df.columns[df.columns.str.contains(r'C107$')][0]
weight_str=df.columns[df.columns.str.contains(r'C105$')][0]

df.replace({gender_str:5,height_str:-9},women_height,inplace=True)
df.replace({gender_str:5,weight_str:-9},women_weight,inplace=True)
df.replace({gender_str:1,height_str:-9},men_height,inplace=True)
df.replace({gender_str:1,weight_str:-9},men_weight,inplace=True)

# w##bmi string 으로 저장하기.
bmi_str=df.columns[df.columns.str.contains(r'bmi$')][0]

# 원래있던 bmi column 없애기
df.drop(columns=[bmi_str],inplace=True)

# BMI = 체중(kg) / 키(m)의 제곱
df=df.assign(
    bmi=lambda x: x[weight_str]/((x[height_str]/100)**2)
)
#새로운 bmi로 대체
df=df.rename({'bmi':bmi_str},axis='columns')
```
